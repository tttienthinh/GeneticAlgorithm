```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 100, 1_000)
close = np.sin(time)
```

```python
plt.plot(time, close)
```

```python
df = pd.DataFrame()
```

```python
df["close"] = close
df["high"] = close + np.random.random(len(time))
df["low"] = close - np.random.random(len(time))
df["volume"] = np.random.random(len(time))
```

```python
df.to_csv("1-Data creation/0-price.csv", index=False)
```
